#### **6.2 Checking effects of parameters on quality**

```python
plt.scatter(dataset['alcohol'], dataset['quality'], color='red', alpha=0.5)
plt.title('Alcohol vs. Quality')
plt.xlabel('Alcohol')
plt.ylabel('Quality')
plt.grid(True)
plt.show()
```

```python
plt.scatter(dataset['chlorides'], dataset['quality'], color='blue', alpha=0.5)
plt.title('Chlorides vs. Quality')
plt.xlabel('Chlorides')
plt.ylabel('Quality')
plt.grid(True)
plt.show()
```

```python
plt.scatter(dataset['citric acid'], dataset['quality'], color='green', alpha=0.5)
plt.title('Citric Acid vs. Quality')
plt.xlabel('Citric Acid')
plt.ylabel('Quality')
plt.grid(True)
plt.show()
```

```python
plt.scatter(dataset['residual sugar'], dataset['quality'], color='black', alpha=0.5)
plt.title('Residual Sugar vs. Quality')
plt.xlabel('Residual Sugar')
plt.ylabel('Quality')
plt.grid(True)
plt.show()
```

#### **6.3 Checking effects of acids on pH**

```python
plt.scatter(dataset['fixed acidity'], dataset['pH'], color='gold', alpha=0.5)
plt.title('Fixed Acidity vs. pH')
plt.xlabel('Fixed Acidity')
plt.ylabel('pH')
plt.grid(True)
plt.show()
```

```python
plt.scatter(dataset['volatile acidity'], dataset['pH'], color='magenta', alpha=0.5)
plt.title('Volatile Acidity vs pH')
plt.xlabel('Volatile Acidity')
plt.ylabel('pH')
plt.grid(True)
plt.show()
```

```python
plt.scatter(dataset['citric acid'], dataset['pH'], color='brown', alpha=0.5)
plt.title('Citric Acid vs pH')
plt.xlabel('Citric Acid')
plt.ylabel('pH')
plt.grid(True)
plt.show()
```

```python
plt.scatter(dataset['total sulfur dioxide'], dataset['pH'], color='cyan', alpha=0.5)
plt.title('Total Sulfur Dioxide vs pH')
plt.xlabel('Total Sulfur Dioxide')
plt.ylabel('pH')
plt.grid(True)
plt.show()
```

#### **6.4 Correlation of Results**

- #### From the histograms in Section 6.1, we can deduce that;

i. The *amount of sulfur dioxide* found in the samples of red wine **ranges from 0 to 200** with an *average values* which **ranges from 0 to 50** and with *few samples* having sulfur dioxide **above 150**

ii. The *pH of the samples* of red wine **ranges from 0 to a little over 4.0** with *average values* **ranging from 3.2 to 3.4** and and *few samples* having pH which fall **below 3.0** and **above 3.6**

iii.The *amount of chlorides* found in the samples of red wine **ranges from 0 to 0.5** with *most samples* containing chlorides **ranging from 0 to 0.2** and only *a few samples* with clorides **above 0.2**

iv. The samples of red wine were found to have **ranging densities** with *most samples* having densities **between 0.94 and 1**

- #### From the scatter diagrams in Section 6.2, we can deduce that;

i. The *__quality__* of the red wine **increases** with an **increases** in its *alcohol content*

ii. The *__quality__* of the red wine **increases** with a **decrease** in its *chloride content*

iii. The *__quality__* of the red wine **increases** across values of *citric content* but more noticeable increase in quality for values **between 0.2 and 0.8**

iv.The *__quality__* of the red wine **increases** with a **decrease** in its *residual sugar content*


- #### From the scatter diagrams in Section 6.3, we can deduce that;

i. An **increase** in the *fixed acidity* **increases** the *pH* and makes it __*less acidic*__.

ii. The *pH* stays in the **range from 2.8 to 3.8** with **increase** in *volatile acidity* with only *__few deviations__*.

iii. An **increase** in the *citric acid* **decreases** the *pH* and makes it __*more acidic*__.

iv. An **increase** in the *sulfur dioxide* content reduces the *pH* range **from 2.8 to 3.8**  to a *pH* range of **3.2 to 3.4** with a *__few deviations__*
